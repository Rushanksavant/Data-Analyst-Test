# Using classifier pipeline on cleaned data


```python
import pandas as pd
```

```python
# making a list of sentiments for entries in data.Review
def add_sentiment_info(data, data_column):
    sentiment= []
    confidence= []

    for i in data_column:
        i = i[:512]
        result = sentiment_analysis(i)[0]
        sentiment.append(result['label'])
        confidence.append(result['score'])
        
    # Using DataFrame.insert() to add a column
    data.insert(1, "Sentiment", sentiment, True)
    data.insert(2, "Confidence Score", confidence, True)
    
    del sentiment
    del confidence
```

### 1. null location entries

```python
df= pd.read_csv("../input/test-data/data_null_location.csv")
df.head()
```

```python
add_sentiment_info(df, df["Review"])
df.head()
```

```python
df.to_csv('sentiment_info_null_loc.csv',index=False)
```

### 2. non-null location entries

```python
df1= pd.read_csv("../input/test-data/data_non_null_location.csv")
df1.head()
```

```python
add_sentiment_info(df1, df1["Review"])
df1.head()
```

```python
df1.to_csv('sentiment_info_non_null_loc.csv',index=False)
```
