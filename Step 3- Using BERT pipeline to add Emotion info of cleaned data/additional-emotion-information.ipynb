```python
from transformers import pipeline
nlp = pipeline("zero-shot-classification")
```

```python
# Testing
nlp("I am a bit excited by my grades.", ["upset", "satisfied", "anger", "happy"])
```

```python
z= nlp("I am a bit excited by my grades.", ["upset", "satisfied", "anger", "happy"])
z['labels'][:2]
```

# Adding emotion information to dataset

```python
def add_emo_info(df, df_column):
    emo= []
    conf= []
    emo1= []
    conf1= []
    for i in range(len(df)):
        emo_info= nlp(df_column[i], ["upset", "satisfied", "anger", "happy"])
        emo.append(emo_info['labels'][0])
        conf.append(emo_info['scores'][0])
        emo1.append(emo_info['labels'][1])
        conf1.append(emo_info['scores'][1])
        
    df.insert(3, "Emotion1", emo, True)
    df.insert(4, "Emotion_conf1", conf, True)
    df.insert(5, "Emotion2", emo1, True)
    df.insert(6, "Emotion_conf2", conf1, True)
    
    del emo
    del conf
    del emo1
    del conf1
```
