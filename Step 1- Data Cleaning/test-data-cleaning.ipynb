# Data Cleaning
A lot of location data is missing, sorting the data having null location values and data with non-null location values, also deleting the rows with no review entries

```python
# Data Cleaning



# Seperating null and non-null location data
nulls=[]
non_nulls=[]
for i in range(len(data)):
    if type(data["Location"][i]) is str:
        non_nulls.append(i) #(i) because builtin_function_or_method' object is not subscriptable
        
    else:
        nulls.append(i) #(i) because builtin_function_or_method' object is not subscriptable
        
# Defining a function to make dataframe with speific index of data        
def sep_data(data_list):
    combined_csv = pd.concat([pd.read_csv('../input/test-data/Data Analyst - Test Data.csv', skiprows= i+1, nrows=1, header=None) for i in data_list])
    combined_csv= combined_csv.set_index(i for i in range(len(data_list)))
    combined_csv.columns = data.columns
    
    #removing the entries with null reviews
    combined_csv.dropna(subset = ["Review"], inplace=True)
    combined_csv= combined_csv.reset_index(drop=True) # index reset

    return combined_csv

# Data containing null Location entries
data_nulls= sep_data(nulls)

# Data containing non-null Location entries
data_non_nulls= sep_data(non_nulls)

# Displaying datas
print("Data with null Location entries:")
display(data_nulls)

print("Data with non-null Location entries:")
display(data_non_nulls)
```

```python
data_nulls= data_nulls.drop(columns= ['Unnamed: 3'])
data_non_nulls= data_non_nulls.drop(columns= ['Unnamed: 3'])
```
